# Importing libraries & read data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
summer = pd.read_csv('data/summer.csv')
```

```python
winter = pd.read_csv('data/winter.csv')
```

```python
conteries = pd.read_csv("data/dictionary.csv")
```

```python
summer=summer.merge(conteries, left_on='Country',right_on='Code' ,how='left').drop(columns=['Code','Population','GDP per Capita'])
                            
summer.head()
```

```python
summer.rename(columns={"Country_x":'Code',"Country_y":'Country'},inplace=True)
```

```python
summer
```

```python
summer['Athlete']=summer['Athlete'].str.split(", ").str[::-1].str.join(" ")
summer['Athlete']=summer['Athlete'].str.title()
```

```python
summer
```

```python
summer.isna().sum()
```

```python
winter=winter.merge(conteries, left_on='Country',right_on='Code' ,how='left').drop(columns=['Code','Population','GDP per Capita'])
                            
winter.head()
```

```python
winter.rename(columns={"Country_x":'Code',"Country_y":'Country'},inplace=True)
```

```python
winter
```

```python
winter.isna().sum()
```

```python
olympics=pd.concat([summer , winter], keys=['summer','winter'],names=['edition']).reset_index().drop(columns='level_1')
olympics
```

```python
olympics.isna().sum()
```

```python
olympics[olympics['Country'].isna()]
```

```python
olympics[olympics['Country'].isna()].Code.value_counts()
```

```python
old_code=olympics[olympics['Country'].isna()].Code.value_counts().index
old_code
```

```python
mapper=pd.Series(index=old_code,name='Country',data=[ 'Soviet Union', 'East Germany', 'Romania', 'West Germany', 'Czechoslovakia',
                                              'Yugoslavia', 'Unified Team', 'Unified Team of Germany', 'Mixed teams', 
                                              'Serbia', 'Australasia', 'Russian Empire', 'Montenegro', 'Trinidad and Tobago',
                                              'Bohemia', 'West Indies Federation', 'Singapore',
                                              'Independent Olympic Participants'])
```

```python
mapper
```

```python
missing_indices=olympics.loc[olympics['Country'].isna()].index
missing_indices
```

```python
olympics.loc[missing_indices,"Code"].map(mapper)
```

```python
olympics['Country'].fillna(olympics['Code'].map(mapper),inplace=True)
```

```python
olympics
```

```python
olympics[olympics['Code'].isna()]
```

```python
olympics.dropna(subset='Code',inplace=True)
```

```python
olympics
```

```python
olympics.isna().sum().sum()
```

```python
olympics.reset_index(drop=True , inplace=True)
```
