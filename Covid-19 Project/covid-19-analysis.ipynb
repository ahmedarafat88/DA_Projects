```python
track_1 = go.Bar(
            x = data_continent.index, 
            y = data_continent['total_cases'],
            name = 'Total Cases',
            )

track_2 = go.Bar(
            x = data_continent.index, 
            y = data_continent['new_cases'],
            name = 'New Cases',
            )
track_3 = go.Bar(
            x = data_continent.index, 
            y = data_continent['total_deaths'],
            name = 'Total Deaths',
            )
track_4 = go.Bar(
            x = data_continent.index, 
            y = data_continent['new_deaths'],
            name = 'New Deaths ',
            )
data=[track_1 , track_2 , track_3 , track_4]

layout = dict( title = 'Data Continent By (Bar)',title_x=.5 , title_font=dict(size=30), width=900,
    height=600)

fig = dict(data=data , layout =layout)
iplot(fig)
```

```python
track_01 = go.Scatter(
            x = data_continent.index, 
            y = data_continent['total_cases'],
            name = 'Total Cases',
            )

track_02 = go.Scatter(
            x = data_continent.index, 
            y = data_continent['new_cases'],
            name = 'New Cases',
            )
track_03 = go.Scatter(
            x = data_continent.index, 
            y = data_continent['total_deaths'],
            name = 'Total Deaths',
            )
track_04 = go.Scatter(
            x = data_continent.index, 
            y = data_continent['new_deaths'],
            name = 'New Deaths ',
            )
data=[track_01 , track_02 , track_03 , track_04]

layout = dict( title = 'Data Continent By (Scatter)',title_x=.5 , title_font=dict(size=30), width=900,
    height=600)

fig = dict(data=data , layout =layout)
iplot(fig)
```
