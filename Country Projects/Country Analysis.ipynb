```python
# The ten most populous countries 
top_10=data.groupby(['Country','Population'])['Population'].max().sort_values(ascending=False).index[0:10] 
top_10             
```

```python
country=[]
population=[]
for i in top_10:
    print(i[0],i[1])
    country.append(i[0])
    population.append(i[1])  
plt.figure(figsize=(20,10))    
plt.bar(x=country,height=population,color='red',width=0.5)
plt.xlabel('Country',fontsize=25)
plt.xticks(rotation=45,fontsize=20)
plt.ylabel('Population',fontsize=25)
plt.yticks(rotation=45,fontsize=20)
plt.title('The ten most populous countries',fontsize=30)
```

```python
# The ten least populous countries
low_10=data.groupby(['Country','Population'])['Population'].max().sort_values(ascending=True).index[0:10] 
low_10    
```

```python
count=[]
popul=[]
for i in low_10:
    print(i[0],i[1])
    count.append(i[0])
    popul.append(i[1])  
plt.figure(figsize=(20,10))    
plt.bar(x=count,height=popul,color='red',width=0.5)
plt.xlabel('Country',fontsize=25)
plt.xticks(rotation=45,fontsize=20)
plt.ylabel('Population',fontsize=25)
plt.yticks(rotation=45,fontsize=20)
plt.title('The ten least populous countries',fontsize=30)
```

# Check outliers

```python
plt.figure(figsize=(10,5))
plt.xlabel('Population', fontsize=15)
plt.title("Check outliers of Population",fontsize=20)
sns.boxplot(data['Population'],color='r')
plt.ylabel('Value',fontsize=20)
plt.xlabel('Population',fontsize=20)
plt.grid(True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# remove outliers
q75,q25 = np.percentile(data['Population'],[75,25])
IQR = q75 - q25
min_out = q25 - 1.5 * IQR
max_out = q75 + 1.5 * IQR
print("q75= ",q75 ,"\n","q25= ", q25 ,"\n" ,"IQR= " ,IQR ,"\n","min_out= ", min_out ,"\n","max_out= ", max_out)
```

```python
# remove outlirers
data.Population[data['Population']> max_out]=max_out
data.Population[data['Population']< min_out]=min_out
sns.boxplot(data['Population'])
plt.title('Remove outliers of Population',fontsize=20)
plt.ylabel('Value',fontsize=20)
plt.xlabel('Population',fontsize=20)
plt.grid(True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```
